```python
import pandas as pd
import json
import cobra
import numpy as np
```

```python
# read biolog base composition file, original plata table, and mapping file for Plata names -> seed IDs
biolog_names_to_seed = pd.read_csv('../data/biolog_names_to_seed.csv',sep='\t')
plata_biolog_data = pd.read_csv('../data/plata_biolog_raw.csv',sep=',',index_col=0)
```

```python
# replace plata metabolite names with seed IDs
plata_biolog_data = plata_biolog_data.rename(dict(zip(biolog_names_to_seed['plata_name'],biolog_names_to_seed['seed_id'])),axis='columns')
# remove the NA columns (did not map due to criteria listed in manuscript)
plata_biolog_data = plata_biolog_data.loc[:,~plata_biolog_data.columns.isnull()]
plata_biolog_data.columns = plata_biolog_data.columns + '_e'
```

```python
# filter organisms by having growth in at least 20 conditions total
growth_threshold = 10 # as in Plata paper. These are relative colorimetric units. See paper for details.
min_growth_conditions = 20
plata_biolog_data = plata_biolog_data.loc[(plata_biolog_data > growth_threshold).sum(axis=1) > min_growth_conditions,]
```

```python
plata_biolog_data = plata_biolog_data > 10
plata_biolog_data
```

```python
plata_biolog_data.to_csv('../data/plata_thresholded.csv',sep='\t')
```
